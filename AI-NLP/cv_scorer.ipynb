```python
def score_cv_for_role(cv_text: str, job_description: str) -> str:
    """
    Generates a suitability score and explanation from AI based on the CV text and job description.
    """
    prompt = f"""
You are an experienced Human Resources and Technical Recruitment Specialist.
Below, a JOB DESCRIPTION and a CANDIDATE CV will be provided.

Your Task:
1) Score the candidate's suitability for this position between 0-100.
2) List the strong matching aspects (skills, experience, education) as bullet points.
3) List missing or weak aspects as bullet points.
4) Add a short general comment at the end.

Please answer in the following format:

Suitability Score: XX/100

Key Strengths:
- ...

Weaknesses / Gaps:
- ...

General Comment:
- ...

Notes:
- Ensure the response is in ENGLISH.
- Please evaluate based ONLY on the provided job description and CV.
- Do not add fabricated information or hallucinations.

JOB DESCRIPTION:
\"\"\"{job_description}\"\"\"

CANDIDATE CV TEXT:
\"\"\"{cv_text}\"\"\"
"""
    
    response = model.generate_content(prompt)
    return response.text
```

```python
import numpy as np

genai.configure(api_key="AIzaSyCLXnTDnOL43oN1X5IF_0uBD4dRxQyuHPs")

EMBEDDING_MODEL = "models/text-embedding-004"
GEN_MODEL = "models/gemini-flash-latest"

gen_model = genai.GenerativeModel(GEN_MODEL)

def split_into_chunks(text, max_chars=800):
    
    chunks = []
    current = ""
    for line in text.splitlines():
        if len(current) + len(line) + 1 <= max_chars:
            current += line + "\n"
        else:
            if current.strip():
                chunks.append(current.strip())
            current = line + "\n"
    if current.strip():
        chunks.append(current.strip())
    return chunks

def embed_text(text: str):
    
    res = genai.embed_content(
        model=EMBEDDING_MODEL,
        content=text,
        task_type="retrieval_document"
    )
    emb = res["embedding"]
    return np.array(emb, dtype=float)

# 1) Chunk the CV
cv_chunks = split_into_chunks(pdf_text, max_chars=800)

# 2) Calculate embedding for each chunk
cv_chunk_embeddings = [embed_text(chunk) for chunk in cv_chunks]

print(f"Total number of chunks: {len(cv_chunks)}")
print("First chunk example:\n", cv_chunks[0][:300], "...")
```

```python
import numpy as np
import google.generativeai as genai

genai.configure(api_key="AIzaSyCLXnTDnOL43oN1X5IF_0uBD4dRxQyuHPs")

EMBEDDING_MODEL = "models/text-embedding-004"
GEN_MODEL = "models/gemini-flash-latest" 

gen_model = genai.GenerativeModel(GEN_MODEL)

def embed_text(text: str):
    
    res = genai.embed_content(
        model=EMBEDDING_MODEL,
        content=text,
        task_type="retrieval_document"
    )
    emb = res["embedding"]
    return np.array(emb, dtype=float)
```

```python
cv_chunk_embeddings = [embed_text(chunk) for chunk in cv_chunks]

print("Number of chunks:", len(cv_chunks))
print("Length of 1st chunk:", len(cv_chunk_embeddings[0]))
```

```python
def cosine_similarity(a: np.ndarray, b: np.ndarray) -> float:
    if np.linalg.norm(a) == 0 or np.linalg.norm(b) == 0:
        return 0.0
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))

def retrieve_relevant_chunks(question: str, top_k: int = 3):
    q_emb = embed_text(question)
    scores = []

    for i, emb in enumerate(cv_chunk_embeddings):
        sim = cosine_similarity(q_emb, emb)
        scores.append((i, sim))

    scores.sort(key=lambda x: x[1], reverse=True)

    top_chunks = []
    for idx, sim in scores[:top_k]:
        top_chunks.append((cv_chunks[idx], sim))

    return top_chunks
```

```python
# Mini test:
q = "What are the candidate's technical strengths?"
for i, (chunk, sim) in enumerate(retrieve_relevant_chunks(q), start=1):
    print(f"\n--- Chunk #{i} (similarity={sim:.3f}) ---\n")
    print(chunk[:400], "...")
```

```python
# Now we take those most relevant chunks and provide them to Gemini as "context":
def rag_strengths_weaknesses(question: str, top_k: int = 3) -> str:
    top_chunks = retrieve_relevant_chunks(question, top_k=top_k)
    context_texts = [chunk for (chunk, score) in top_chunks]
    context = "\n\n---\n\n".join(context_texts)

    prompt = f"""
You are an experienced human resources specialist.
Below you see selected parts from a candidate's CV that are most relevant to the question.
Try to generate an answer without going outside the scope of these parts.

Question:
\"{question}\"

Relevant CV Parts:
\"\"\"{context}\"\"\"

Please answer in English in the following format:

1) Strengths:
- ...

2) Weaknesses:
- ...

3) General Evaluation (short 2–4 sentences):
- ...

Note:
- Do not fabricate things that are not explicitly mentioned in the CV.
- If you are unsure, write "It is not clearly understood from the CV".
"""
    response = gen_model.generate_content(prompt)
    return response.text
```

```python
question = "What are the candidate's technical and general strengths and weaknesses?"
result = rag_strengths_weaknesses(question, top_k=3)
print(result)
```

```python
def retrieve_relevant_chunks_for_job(job_description: str, top_k: int = 3):
    job_emb = embed_text(job_description)
    
    scores = []
    for i, emb in enumerate(cv_chunk_embeddings):
        sim = cosine_similarity(job_emb, emb)
        scores.append((i, sim))
    
    scores.sort(key=lambda x: x[1], reverse=True)
    
    top_chunks = []
    for idx, sim in scores[:top_k]:
        top_chunks.append((cv_chunks[idx], sim))
    
    return top_chunks
```

```python
job_description = """
Junior Data Analyst pozisyonu için;
- Çok iyi derece Python
- Çok iyi derece SQL
- Temel istatistik ve veri analizi
- Tercihen Machine Learning veya NLP bilgisi
- Takım çalışması ve iletişim becerileri
aranmaktadır.
"""

for i, (chunk, sim) in enumerate(retrieve_relevant_chunks_for_job(job_description), start=1):
    print(f"\n--- Most relevant chunk #{i} (similarity={sim:.3f}) ---\n")
    print(chunk[:400], "...")
```

```python
def rag_job_fit_score(cv_text: str, job_description: str, top_k: int = 3) -> str:
    
    top_chunks = retrieve_relevant_chunks_for_job(job_description, top_k=top_k)
    context_texts = [chunk for (chunk, score) in top_chunks]
    context = "\n\n---\n\n".join(context_texts)

    prompt = f"""
You are an experienced Human Resources and Technical Recruitment Specialist.

Below, a JOB DESCRIPTION and selected parts from a CANDIDATE'S CV that most resemble
the job description will be provided.

Your Task:
1) Score the candidate's suitability for this position between 0–100.
2) List the strong matching aspects as bullet points.
3) List missing or weak aspects as bullet points.
4) Add a short general comment.

Please answer in the following format:

Suitability Score: XX/100

Key Strengths:
- ...

Weaknesses / Gaps:
- ...

General Comment:
- ...

Ensure the response is in ENGLISH. Do not fabricate things that are NOT CLEAR in the CV or job description.

JOB DESCRIPTION:
\"\"\"{job_description}\"\"\"

SELECTED RELEVANT PARTS FROM CANDIDATE CV:
\"\"\"{context}\"\"\"
"""

    response = gen_model.generate_content(prompt)
    return response.text
```
